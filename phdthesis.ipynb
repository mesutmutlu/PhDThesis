```python
gii_corr_cols = gii_cols.copy()
gii_corr_cols.remove('Year')
gii_corr_cols.remove('Country')
#print(gii_corr_cols)
sdg_corr_cols = [i for i in sdg_resume.columns.tolist() if i.startswith('SDG')]
#print(sdg_corr_cols)
def o_corr(df, deps):
    a = []
    for d in deps:
        sdg_corr = df[~df[d].isin([np.inf, np.nan, -np.inf, 0])][[d]+sdg_corr_cols].corr()[[d]]
        gii_corr = df[~df[d].isin([np.inf, np.nan, -np.inf, 0])][[d]+gii_corr_cols].corr(method='spearman')[[d]]
        
        df_corr = pd.concat([sdg_corr,gii_corr], axis = 0)
        #print(df_corr)
        #print(df[~df[d].isin([np.inf, np.nan, -np.inf, 0])].corr()[[d]])
        a.append(df_corr.drop(index=[d]))
        #a.append(df[~df[d].isin([np.inf, np.nan, -np.inf, 0])].corr()[[d]])
    #print(a)  
    #return pd.concat(a, axis = 1).drop(index=lines_to_drop_from_corr).sort_index(axis=1).round(2)
    return pd.concat(a, axis = 1).drop(index=lines_to_drop_from_corr, errors='ignore').round(2)
#print('df_final_2:',df_final_2.shape)
#o_corr(df_final_2,l_Deps)

def corr_test(dfi, samplesize):
    dfi_test = pd.DataFrame()
    for i in range(0,1000):
        df_sample = o_corr(dfi.sample(samplesize),l_Deps)
        #print(b)
        #print('aaaaa',df_sample)
        if dfi_test.empty:
            dfi_test = df_sample
        else:
            dfi_test = dfi_test.astype(str) + ',' + df_sample.astype(str)
        #print('bbbbbb',dfi_test)
    dfi_test = dfi_test.applymap(lambda x: np.fromstring(x, dtype=float, sep=','))
    return dfi_test
```
